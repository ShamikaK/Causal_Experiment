Control Group

```python
control_sent = pd.read_csv('/content/drive/Shareddrives/BA-830-Team/members_Control_group_sent_Apr_21_2021.csv')
control_click = pd.read_csv('/content/drive/Shareddrives/BA-830-Team/members_Control_group_click_activity_Apr_21_2021.csv')
control_open = pd.read_csv('/content/drive/Shareddrives/BA-830-Team/members_Control_group_opened_Apr_21_2021.csv')
control_click1 = pd.read_csv('/content/drive/Shareddrives/BA-830-Team/members_Control_group_Copy_01__click_activity_Apr_21_2021.csv')
control_open1 = pd.read_csv('/content/drive/Shareddrives/BA-830-Team/members_Control_group_Copy_01__opened_Apr_21_2021.csv')
```

```python
control_sent.head()
```

```python
control_sent.shape
```

```python
control_sent = control_sent[['Email Address']]
control_sent.head()
```

```python
control_click = control_click[['Email Address', "Clicks"]]
control_click.head()
```

```python
control_open = control_open[['Email Address', "Opens"]]
control_open.head()
```

```python
control_open1 = control_open1[['Email Address', "Opens"]]
control_click1 = control_click1[['Email Address', "Clicks"]]
```

```python
control = control_sent.merge(control_click, how = 'left', on = 'Email Address').merge(control_open, how = 'left', on = 'Email Address').merge(control_click1, how = 'left', on = 'Email Address').merge(control_open1, how = 'left', on = 'Email Address')
control.head()
```

```python
control.fillna(0, inplace = True)
control.head()
```

```python
control['Clicks'] = control['Clicks_x'] + control['Clicks_y']
control['Opens'] = control['Opens_x'] + control['Opens_y']
```

```python
control = control[['Email Address', "Clicks", "Opens"]]
control.head()
```
